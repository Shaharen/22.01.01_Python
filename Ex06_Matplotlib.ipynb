### 선그래프

```python
# plot함수 - 일직선으로 되어있는 Lineplot 그리기
y = [2,4,6]
plt.plot(y) # 그래프 그리기
plt.show()  # 그래프 출력
```

```python
# x축 설정하여 그래프 그리기
x = [1,2,3]
y = [2,4,6]

plt.plot(x,y)
plt.show()
```

```python
# line style
# 그려지는 라인의 스타일 지정
x = [1,2,3]
y = [2,4,6]

plt.plot(x,y,linestyle='--')
plt.show()
```

```python
x = [2, 4, 6, 8]
y = [10, 13, 16, 20]

y2 = [10, 20, 16, 13]
plt.plot(x,y)
plt.plot(x,y2, ls = ':')  # 점선 표시
plt.show()
```

```python
# marker
plt.plot(x,y,marker='o')
plt.show()
```

```python
# linewidth , lw로 줄여씀
# 라인 두께 지정
plt.plot(x, y, linewidth = '5')
plt.show()
```

```python
plt.plot(x, y, color = 'r')
plt.plot(x, y, color = '#0000ff')
plt.show()
```

```python
# Marker color
# markerfacecolor 로 마커 색 변경 mfc 로 줄여씀
plt.plot(x, y, color = '#039BE5', marker = 'o' , markerfacecolor = 'r')
plt.show()
```

```python
# 한 영역 안에 여러 그래프 그리기
x = [1,2,3,4]
y = [2,4,6,8]
z = [3,6,9,12]

# 그래프에 이름 주기 label
# 준 이름에 따라 범례 출력 legend
plt.plot(x, color = 'r', label = 'x')
plt.plot(y, color = 'k', label = 'y')
plt.plot(z, color = 'm', label = 'z')
plt.legend(loc = 'center')
# best ( 기본값 )
# 	upper right
# 	upper left
# 	lower left
# 	lower right
# 	right
# 	center left
# 	center right
# 	lower center
# 	upper center
# 	center
plt.show()
```

#### 차대차 사건 중 죽거나 다친 사람이 많은 발생지 시도를 알아보고 시각화

```python
# 1. 차대차 사건 중 
# 2. (1) 죽거나 다친 사람(사상자)이 많은  (2) 발생지 시도
# 3. 시각화
```

```python
# 1. 차대차 사건 분류
acc = data[data['사고유형_대분류'] == '차대차']
acc
```

```python
# 2
acc['발생지시도'].unique()
```

```python
acc['사상자수']
```

```python
# 그룹화 : groupby( '그룹화 시킬 컬럼' ).연산

acc_g = acc.loc[:,['사상자수','발생지시도']].groupby('발생지시도').sum()
acc_g
```

```python
# 3 시각화
x = acc_g.index
y = acc_g['사상자수']
# 그래프의 공간 크기 설정 figsize = (가로, 세로)
# 옵션의 가장 위에 위치해야 적용됨
plt.figure(figsize = (10, 5), facecolor = (234/256,171/256,91/256))

plt.title('차vs차 교통사고의 사상자 수' ,color = (0,192/256,0))
plt.xlabel('지역', color = (1,0,1))
plt.ylabel('사상자 수' , rotation = 0 , color = (1, 1, 0))

plt.bar(x,y, color = (25/256,25/256,112/256))
plt.show()
```
